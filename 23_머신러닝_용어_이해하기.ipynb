#1. 머신러닝 모델의 평가
검증용 데이터는 모델의 성능을 평가하기 위한 용도가 아니라, 모델의 성능을 조정하기 위한 용도(과적합이 되고 있는지 판단, 하이퍼파라미터의 조정을 위한 용도)

하이퍼파라미터란 값에 따라 모델의 성능에 영향을 주는 매개변수

가중치, 편향과 같은 학습을 통해 바뀌어져가는 변수를 매개변수라고 함

하이퍼파라미터는 사용자가 직접 정해줄 수 있는 변수(경사 하강법에서 학습률(learning rate), 딥러닝에서는 은닉층의 수, 뉴런의 수, 드롭아웃 비율)

훈련용 데이터로 훈련을 모두 시킨 모델은 검증용 데이터를 사용하여 정확도를 검증하며 하이퍼파라미터를 튜닝(tuning)함.

모델의 매개변수는 검증용 데이터로 정확도가 검증되는 과정에서 점차 검증용 데이터에 맞춰가기 시작

하이퍼파라미터 튜닝이 끝났다면 검증용 데이터로 모델을 평가하는 것은 적합하지 않음(검증용 데이터에 대해서도 일정 부분 최적화가 되어있기 때문). 모델에 대한 평가는 모델이 아직까지 보지 못한 데이터(=테스트 데이터)를 가지고 모델의 진짜 성능을 평가.

검증 데이터와 테스트 데이터를 나눌 만큼 데이터가 충분하지 않다면 k-폴드 교차 검증

#6. Overfitting & Underfitting

Overfitting이란 훈련 데이터를 과하게 학습한 경우

-> 훈련 데이터에 대해서만 과하게 학습하면 테스트 데이터나 실제 서비스 데이터에 대해서는 정확도가 좋지 않음

과적합 상황에서는 훈련 데이터에 대한 호차는 낮지만, 테스트 데이터에 대해서 오차가 높아짐

image.png

X축의 epoch는 전체 훈련 데이터에 대한 훈련 횟수 의미

3~4회를 넘어가게 되면 과적합 발생

과적합 방지를 위해 테스트 데이터의 성능이 낮아지기 전에 훈련을 멈추는 것이 바람직한데, 테스트 데이터의 성능이 올라갈 여지가 있음에도 훈련을 덜한 상태를 반대로 Underfitting이라고 함.
