```python
import pandas as pd # csv 로드
from sklearn.model_selection import train_test_split # data 8:2
from sklearn.preprocessing import LabelEncoder # 텍스트 -> 숫자형 바꿀때
from torchtext.data.utils import get_tokenizer # 토크나이저
from torchtext.vocab import build_vocab_from_iterator # 단어사전
 # 단어사전
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import accuracy_score
import warnings
import torchtext
```

```python
# 데이터 로딩
df = pd.read_csv('/content/nsmc/ratings_train.txt', delimiter='\t')
df.columns = ['id', 'document', 'label']
```

```python
# NaN 값을 빈 문자열로 변환
df['document'] = df['document'].astype(str)
```

```python
# 데이터 분할
train_df, valid_df = train_test_split(df, test_size=0.2, shuffle=True, random_state=42)
train_df
```

```python
from konlpy.tag import Mecab
from torchtext.data.utils import get_tokenizer

mecab = Mecab()

def custom_tokenizer(text):
  return mecab.morphs(text)
```

```python
# 유지보수용
tokenizer = get_tokenizer(custom_tokenizer)
```

```python
def yield_tokens(data_iter):
  for text in data_iter:
    yield tokenizer(text)

vocab = build_vocab_from_iterator(yield_tokens(df['document']), specials=['<unk>'])

vocab.set_default_index(vocab['<unk>'])
```

```python
# 데이터셋 클래스 : 데이터를 불러오거나 정의, 전처리하는 클래스함수
# 유지보수
class ReviewDataset(Dataset):
  def __init__(self, dataframe, vocab, tokenizer):
    self.data = dataframe # 데이터
    self.vocab = vocab # 단어리스트를 보캡의 인덱스 이용해서 숫자로 컨버팅
    self.tokenizer = tokenizer # mecab morphs 단어단위로 자르는

  def __len__(self):
    return len(self.data)

  def __getitem__(self, idx): # 데이터를 불러올 때 반응하는 함수 argument로
                              #몇 번째 데이터를 불러올건지 인덱스를 받음
    text = self.data.iloc[idx]['document'] # 받은 행 인덱스 (몇번째 데이터)로 데이터 선택
    label = self.data.iloc[idx]['label'] # 받은 행 인덱스 (몇 번째 데이터)로 정답 선택
    tokenized_text = self.tokenizer(text) # 단어 쪼개는 것
    numerical_text = torch.tensor(self.vocab(tokenized_text))
    return numerical_text, torch.tensor(label)
```

```python
train_dataset = ReviewDataset(train_df, vocab, tokenizer)
valid_dataset = ReviewDataset(valid_df, vocab, tokenizer)
```

```python
import torch

sequence_batch = [
    torch.tensor([1,2,3]),
    torch.tensor([1,2,]),
    torch.tensor([1])
]
torch.nn.utils.rnn.pad_sequence(sequence_batch, batch_first=True)
```

```python
def collate_fn(batch): # 콜백함수
  texts, labels = zip(*batch)
  texts = torch.nn.utils.rnn.pad_sequence(texts, batch_first=True)
  labels = torch.stack(labels)

  return texts, labels

train_loader = DataLoader(train_dataset, batch_size=2, shuffle=True, collate_fn=collate_fn)
valid_loader = DataLoader(valid_dataset, batch_size=2, collate_fn=collate_fn)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from sklearn.metrics import accuracy_score

# RNN 모델 정의
class RNNClassifier(nn.Module):
  def __init__(self, vocab_size, embed_size, hidden_size, num_layers=1):
    super(RNNClassifier, self).__init__()
    self.embedding = nn.Embedding(vocab_size, embed_size)
    self.rnn = nn.RNN(embed_size, hidden_size, num_layers, batch_first=True)
    self.fc = nn.Linear(hidden_size, 1) # 이진분류를 위해 output_size를 1로 설정

  def forward(self,x):
    x = self.embedding(x)
    h, _ = self.rnn(x)
    out = self.fc(h[:, -1, :])
    return out
```

```python
# 하이퍼파라미터 설정
embed_size = 100
hidden_size = 128
output_size = 1
num_layers = 1
learning_rate = 0.001
num_epochs = 5

# 모델, 손실함수, 옵티마이저 초기화
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = RNNClassifier(len(vocab), embed_size, hidden_size, num_layers).to(device)
criterion = nn.BCEWithLogitsLoss() # 이진 분류를 위한 손실함수
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

```python
# 훈련 함수
def train():
  model.train()
  for epoch in range(num_epochs):
    total_loss = 0
    for inputs, labels in train_loader:
      inputs, labels = inputs.to(device), labels.to(device, dtype=torch.float) # 레이블을 float으로 변환
      optimizer.zero_grad()
      outputs = model(inputs).squeeze() # 차원축소
      loss = criterion(outputs, labels)
      loss.backward()
      optimizer.step()
      total_loss += loss.item()
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss/len(train_loader):.4f}')

# 검증 함수
def evaluate(loader):
  model.eval()
  all_labels = []
  all_preds = []
  with torch.no_grad():
    for inputs, labels in loader:
      inputs, labels = inputs.to(device), labels.to(device, dtype=torch.float)
      outputs = model(inputs).squeeze()
      preds = (torch.sigmoid(outputs) > 0.5).float() # sigmoid와 임계값으로 예측 변환
      all_labels.extend(labels.cpu().numpy())
      all_preds.extend(preds.cpu().numpy())
  return accuracy_score(all_labels, all_preds)

# 모델 훈련 및 검증
train()
valid_accuracy = evaluate(valid_loader)
print(f'Validation Accuracy: {valid_accuracy:.4f}')
```

```python
# 인퍼런스 예제
def infer(text):
    model.eval()
    tokenized_text = tokenizer(text)
    numerical_text = torch.tensor(vocab(tokenized_text)).unsqueeze(0).to(device)
    with torch.no_grad():
        output = model(numerical_text).squeeze()
        predicted = torch.sigmoid(output).item() > 0.5
        return int(predicted)

example_text = "아 이거 좋다"
prediction = infer(example_text)
print(f'Predicted label for "{example_text}": {prediction}')
```
